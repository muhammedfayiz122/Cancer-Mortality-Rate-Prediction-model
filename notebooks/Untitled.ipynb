```python
X["County"] = X[X.columns[10]].str.split(",").str[0]
X["State"] = X[X.columns[10]].str.split(",").str[1]
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
X_modified.columns[-2]
```

```python
X_encoded = X_modified
X_encoded[X_encoded.columns[-2]] = encoder.fit_transform(X_encoded[X_encoded.columns[-2]])
X_encoded.head()
```

```python
X_encoded[X_encoded.columns[-1]] = encoder.fit_transform(X_encoded[X_encoded.columns[-1]])
X_encoded.head()
```

```python
#onehot encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('geography', OneHotEncoder(), [32] )],remainder= "passthrough")
```

```python
X_encoded = ct.fit_transform(X_encoded)
X_encoded
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
y = df['TARGET_deathRate']
y.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded[:,1:], y, test_size=0.2, random_state=42)
```

```python
X_test[:]
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
